```r
library(tidyverse)
library(tidyselect)
library(lunar)
library(tidyr)
library(dplyr)
```

## Récuperation des données

```r
df <- readxl::read_excel("data/irsocsd20151_T79JNAIS.xls", 
                         sheet="FM",
                         range="D6:AH581",
                         col_names=as.character(1:31),
                         na="so")
```

```r
df
```

###  Reorganisation des données

```r
df$month <- 1:12
```

```r
df$year <- rep(1968:2015, each=12)
```

```r
df
```

```r
gather(df, "day", "births", 1:31) -> df
df %>% drop_na() -> df
```

```r
df %>% group_by(year) %>% summarize(b=sum(births))
df
```

```r
lunar.phase(as.Date("2019-05-23"), name=TRUE)
```

```r
df <- unite(df, "date", c("year", "month", "day"), sep="-")
```

```r
df$moon <- lunar.phase(as.Date(df$date), name=8, shift=1.5)
```

```r
df
```

```r
print(binom.test(145, 1000, 1/8, alternative="greater", conf.level=0.99))
```
