```python
import numpy as np
import skimage
import math
import matplotlib.pyplot as plt
```

```python
import glob

path_images = glob.glob("../EPITA Cours1/Dataset/*.jpg")
list_images = []
for img in path_images:
    im = skimage.io.imread(img)
    list_images.append(im)
    
len(list_images)
```

```python
plt.imshow(list_images[6])
```

```python
def lbp(image, neighboor=3):
    gray_image = skimage.color.rgb2gray(image)
    output = gray_image.copy()
    val = neighboor//2
    print(val)
    for i in range(val, gray_image.shape[0] - val):
        for j in range(val,  gray_image.shape[1] - val):
            im = gray_image[i-val:i+val+1, j-val:j+val+1]
            ref = im[val, val]
            im = im.flatten()
            im = np.delete(im, len(im) // 2)
            res = 0
            for k in range(len(im)):
                if im[k] >= ref:
                    res += 2**k
                    
            #print(res)
            output[i,j] = res / 255.
    return output
        
    
```

```python
out = lbp(list_images[6])
```

```python
plt.imshow(out, cmap='gray')
```
